# Starting with some data
Starting very simply, let's bring in some data from the _daily_ data set and plot. We'll start with a very basic function that can read a data set as a data frame and plot the Close price over a period of time.

```python
import pr_model

pr_model.plot_symbol_history('AAPL')
```

Next we'll take a look at a given stock's returns over this period of time, on a daily basis. For _AAPL_ it looks something like this:

```python
pr_model.plot_return_timeline('AAPL')
```

On a daily basis, as is expected, the return hovers around some small number thats slightly positive but near zero. Let's bin this data and look at how the data fits into a normal distribution.

```python
pr_model.plot_return_distribution('AAPL')
```

One interesting thing to note here is that with the normal distribution, there's a probability of basically zero once you get out past a few standard deviation, however clearly from the graph those events happen with some frequency. See the 10% drop in a single day and worse on one end, and also the 20% bump in a single day at the other end. Many agree that the using a normal distribution for these purposes is too much of an oversimplification, even despite the Law of Large Numbers and the Central Limit Theorem. The key assumption in the Central Limit Theorem is that the random events being summed are _independent_, which is simply not an assumption that we can make about price movements of various stocks.
